```python
def isInsidePolygon(x, y,xdot, ydot):
    for j in range (len(x)):
        #val = (x[j]-x[j-1])*(xdot-x[j-1])+(y[j]-y[j-1])*(ydot-y[j-1])
        val = (x[j]-x[j-1])*(ydot-y[j-1])-(y[j]-y[j-1])*(xdot-x[j-1])
        #print(val)
        if val < 0:
            return False
    return True
```
