```python
import matplotlib.pyplot as plt
import numpy as np
import math
from random import randint
```

```python
def oneDim(func_comp):
    n=5
    x = [randint(0, 50) for p in range(n)]
    print(x)
    y = [(0) for p in range(n)]
    m=abs(x[1]-x[0])
    ind1=1;
    ind2=0;
    for i in range(0, n, 1):
        for j in range(i+1, n,1):
            if(func_comp(abs(x[i]-x[j]), m)):
                m=abs(x[i]-x[j])
                ind1=i
                ind2=j

    print("value=",m,'  ind1=', ind1,'  ind2=', ind2);
    plt.scatter(x, y)
    x1=[x[ind1], x[ind2]]
    y1=[0,0]
    plt.plot(x1, y1)
    
    for i in range(0, n, 1):
        for j in range(i+1, n,1):
            if(abs(x[i]-x[j]==m)):
                x1=[x[i], x[j]]
                y1=[0,0]
                plt.plot(x1, y1)
                
    
    plt.show()
    
def min_comparison(x, y):
    return x < y
def max_comparison(x, y):
    return x > y
```

```python
def twoDim(func_comp):
    n=5
    x = [randint(0, 50) for p in range(n)]
    y = [randint(0, 50) for p in range(n)]
    print(x)
    print(y)
    m=math.sqrt(pow(x[1]-x[0],2)+pow(y[1]-y[0],2))
    ind1=1
    ind2=0
    for i in range(0, n, 1):
        for j in range(i+1, n,1):
            if func_comp(math.sqrt(pow(x[i]-x[j],2)+pow(y[i]-y[j],2)), m):
                m=math.sqrt(pow(x[i]-x[j],2)+pow(y[i]-y[j],2))
                ind1=i
                ind2=j

    print("value=",m,'  ind1=', ind1,'  ind2=', ind2);
    plt.scatter(x, y)
    x1=[x[ind1], x[ind2]]
    y1=[y[ind1], y[ind2]]
    plt.plot(x1, y1)
    plt.show()
    
```

```python
def threeDim(func_comp):
    n=5
    x = [randint(0, 50) for p in range(n)]
    y = [randint(0, 50) for p in range(n)]
    z = [randint(0, 50) for p in range(n)]
    print(x)
    print(y)
    print(z)
    m=math.sqrt(pow(x[1]-x[0],2)+pow(y[1]-y[0],2)+pow(z[1]-z[0],2))
    ind1=1;
    ind2=0;
    for i in range(0, n, 1):
        for j in range(i+1, n,1):
            if func_comp(math.sqrt(pow(x[i]-x[j],2)+pow(y[i]-y[j],2)+pow(z[i]-z[j],2)), m):
                m=math.sqrt(pow(x[i]-x[j],2)+pow(y[i]-y[j],2)+pow(z[i]-z[j],2))
                ind1=i
                ind2=j

    print("value=",m,'  ind1=', ind1,'  ind2=', ind2);
    
    fig = plt.figure(figsize=(7, 7))
    ax = fig.add_subplot(111, projection='3d')   
    ax.scatter(x, y,z)
    x1=[x[ind1], x[ind2]]
    y1=[y[ind1], y[ind2]]
    z1=[z[ind1], z[ind2]]
    ax.plot(x1, y1,z1)  
    plt.show()
```

```python
while True:
    print('1 - one dimensional')
    print('2 - two dimensional')
    print('3 - three dimensional')
    print('0 - Exit')
    op=input()
    if op=="0":
        break;
    n=5
    x = [randint(0, 30) for p in range(n)]         
    match op:
        case "1":
            print('1 - min')
            print('2 - max')
            op1=input()
            match op1:
                case "1":                    
                    oneDim(min_comparison)
                                        
                case "2":
                    oneDim(max_comparison)
                
            
        case "2":
            print('1 - min')
            print('2 - max')
            op1=input()
            match op1:
                case "1":                    
                    twoDim(min_comparison)
                                        
                case "2":
                    twoDim(max_comparison)
            

        case "3":
            print('1 - min')
            print('2 - max')
            op2=input()
            match op2:
                case "1":                    
                    threeDim(min_comparison)
                case "2":                    
                    threeDim(max_comparison)
               
```
