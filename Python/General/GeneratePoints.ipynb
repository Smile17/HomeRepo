```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
n = 5
color = 'red'
```

```python
for n in range(18,21):
    arr = np.random.rand(2,n)
    fig, ax = plt.subplots(figsize=(5, 5))
    fig.subplots_adjust(0,0,1,1)
    ax.axis("off")
    ax.set_xlim(-0.2, 1.2)
    ax.set_ylim(-0.2, 1.2)
    ax.scatter(arr[0], arr[1], s = 800, color = color)
    plt.savefig('res/' + str(n) + '.png')
```
