```python
import numpy as np
import math
import matplotlib.pyplot as plt
import time

# Обираємо параметри генератора
m1=17
n1 = 5
z1 = 2**10

# Функція, що генерує число з заданим законом розподілу
def Rand(a,b,n): # Звернути увагу на те, від чого залежить генератор
    x=time.perf_counter() # Перший х вибираємо випадково
    r=[]
    for i in range(n+1):
        x=(m1*x+n1)%z1 # 1 крок класичного лінійного конгруентного методу
        r1=x/z1 # 2 крок класичного лінійного конгруентного методу, r1 - рівномірно розподілене число від 0 до 1
        r=r+[round(a+(b-a)*r1,3)] # Занести власну формулу для відповідного розподілу
        # Формула у рядку вище перетворює число від 0 до 1 до такого, що має заданий закон розподілу
        # Всі формули представлені у презентації до лекції 8
    return(r)

# Створюємо числа з заданим розподілом (тут - рівномірний від 0 до 5) у кількості 
R=Rand(0,5,2000)

# Функція, що підсумовує кількість чисел у заданому проміжку
def NumR(a,b,R):
    R=np.array(R)
    return(((a <= R) & (R < b)).sum())

# Функція, що оцінює скільки чисел потрапило в кожен проміжок 
def Distribution(R,N):
    Rmin=min(R)
    Rmax=max(R)
    h=round((Rmax-Rmin)/N,2)
    Levels=[round(Rmin+h*i+h/2,2) for i in range(N)]
    Probab=[round(NumR(Levels[i]-h/2,Levels[i]+h/2,R)/len(R)/h,3) for i in range(N)]
    return(Levels,Probab)

# Функція, що будує густину ймовірності для згенерованої послідовності
def Analyze(R,N):
    D=Distribution(R,N)
    plt.plot(D[0],D[1])
    plt.title('Continuous uniform distribution from '+ str(round(min(R)))+' to '+str(round(max(R))))
    plt.ylabel('Probability density function')
    plt.xlabel('Random number')
    plt.xlim(min(R),max(R))
    plt.ylim(0,max(D[1]))
    plt.grid(True)
    
Analyze(R,100) # 100 - кількість проміжків, на які ділимо для аналізу
```

# Розподіл Коші 

Були обрані саме ці параметри генератора, оскільки для них спостерігається найбільший період

```python
# Обираємо параметри генератора
m1=17
n1 = 5
z1 = 2**10

# Функція, що генерує число з заданим законом розподілу
def Rand(x0,gamma,n): # Звернути увагу на те, від чого залежить генератор (x0 = 0)
    x=time.perf_counter() # Перший х вибираємо випадково
    r=[]
    for i in range(n+1):
        x=(m1*x+n1)%z1 # 1 крок класичного лінійного конгруентного методу
        r1=x/z1 # 2 крок класичного лінійного конгруентного методу, r1 - рівномірно розподілене число від 0 до 1
        r=r+[round(x0+gamma*np.tan(math.pi*r1),3)] # Занести власну формулу для відповідного розподілу
        # Формула у рядку вище перетворює число від 0 до 1 до такого, що має заданий закон розподілу
        # Всі формули представлені у презентації до лекції 8
    return(r)

# Створюємо числа з заданим розподілом (тут - рівномірний від 0 до 5) у кількості 
R=Rand(0,0.5,2000)

# Функція, що підсумовує кількість чисел у заданому проміжку
def NumR(a,b,R):
    R=np.array(R)
    return(((a <= R) & (R < b)).sum())

# Функція, що оцінює скільки чисел потрапило в кожен проміжок 
def Distribution(R,N):
    Rmin=min(R)
    Rmax=max(R)
    h=round((Rmax-Rmin)/N,2)
    Levels=[round(Rmin+h*i+h/2,2) for i in range(N)]
    Probab=[round(NumR(Levels[i]-h/2,Levels[i]+h/2,R)/len(R)/h,3) for i in range(N)]
    return(Levels,Probab)

# Функція, що будує густину ймовірності для згенерованої послідовності
def Analyze(R,N):
    D=Distribution(R,N)
    plt.plot(D[0],D[1])
    plt.title('Continuous uniform distribution from '+ str(round(min(R)))+' to '+str(round(max(R))))
    plt.ylabel('Probability density function')
    plt.xlabel('Random number')
    plt.xlim(min(R),max(R))
    plt.ylim(0,max(D[1]))
    plt.grid(True)
    
Analyze(R,100) # 100 - кількість проміжків, на які ділимо для аналізу
```

як бачимо, розподіл достатньо схожий на теоретичний

і чим більша кількість згенерованих чисел, тим більше наш розподіл стає схожим на теоретичний
