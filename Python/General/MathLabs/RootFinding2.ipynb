```python
import matplotlib.pyplot as plt
import numpy as np
import math

def find_root_secant(a, b, f, eps = 0.001):
    while(abs(b - a) > eps):
        a = a - (b - a) * f(a) / (f(b) - f(a))
        b = b - (a - b) * f(b) / (f(a) - f(b))
    return b

def find_root_tangent_lines(x0, f, eps = 0.001, h = 0.00001, max_iterations = 1000):
    
    for i in range(max_iterations):
        y0 = f(x0)
        y1 = (f(x0 + h) - f(x0 - h)) / (2 * h)
        if abs(y1) < eps:      
            break
        x1 = x0 - y0 / y1            
        if abs(x1 - x0) <= eps:  
            return x1                   
        x0 = x1                        

    return None                        

def plot_func(a, b, f):
    n = 100
    x = np.arange(a, b, (b - a) / n)
    y = f(x)
    plt.plot(x, y)
    plt.plot([a, b], [0, 0], color = 'red', alpha = 0.5, linestyle='dashed')
    
# Variant 7
def func1(x):
    return 3 * x - np.cos(x) - 1

def func2(x):
    return x * x * x + 0.2 * x * x + 0.5 * x - 1.2
```

```python
plot_func(-10, 10, func1)
```

```python
root = find_root_secant(-5, 5, func1)
print("Root:", root)
print("Function value:", func1(root))
```

```python
root = find_root_tangent_lines(5, func1)
print("Root:", root)
print("Function value:", func1(root))
```

```python
plot_func(-5, 5, func2)
```

```python
root = find_root_secant(-5, 5, func2)
print("Root:", root)
print("Function value:", func2(root))
```

```python
root = find_root_tangent_lines(5, func2)
print("Root:", root)
print("Function value:", func2(root))
```
